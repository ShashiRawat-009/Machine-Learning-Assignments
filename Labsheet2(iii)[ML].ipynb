```python
import matplotlib.pyplot as plt 
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [200, 250, 300, 280]
categories = ['A', 'B', 'C', 'A', 'B', 'C']
values = [10, 20, 30, 15, 25, 35]
```

```python
# Line graph for sales over months
plt.figure()
plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# Bar chart showing frequency of categories
plt.figure()
plt.bar(['A', 'B', 'C'], [categories.count('A'), categories.count('B'), categories.count('C')])
plt.title("Category Frequency")
plt.show()
```

```python
# Histogram of values
plt.figure()
plt.hist(values, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram of Values")
plt.show()
```

```python
# Scatter plot of two numerical variables
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
plt.figure()
plt.scatter(x, y, color='red')
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Customize plot with title, labels, and legend
plt.figure()
plt.plot(x, y, label="Line XY")
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# Save plot to file and display inline
plt.figure()
plt.plot(x, y)
plt.title("Plot to Save")
plt.savefig("myplot.png")  # saves as PNG
plt.show()
```
