```python
import pandas as pd
data_dict = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 22],
    "Salary": [50000, 60000, 45000]
}
df = pd.DataFrame(data_dict)
```

```python
print("\nDataFrame:\n", df)
```

```python
# Filter rows where Age > 23
filtered_df = df[df['Age'] > 23]
```

```python
print("\nFiltered DataFrame:\n", filtered_df)
```

```python
 #. Add a new column based on existing column
df['Bonus'] = df['Salary'] * 0.1
```

```python
print("\nDataFrame with Bonus:\n", df)
```

```python
# Group data by a categorical column and calculate average
# Adding Department column
df['Department'] = ['HR', 'IT', 'HR']
grouped = df.groupby('Department').mean(numeric_only=True)
```

```python
print("\nGrouped DataFrame (mean):\n", grouped)
```
